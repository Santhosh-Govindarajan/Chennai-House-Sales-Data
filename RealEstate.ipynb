```python
#Checking the unique values of the columns
for i in df.columns:
    values = df[i].unique()
    print(i," : ",values)
    print("\n")

#-->For a single value there are so many different spelings are there , so we have to clean it
```

```python
#Cleaning the spellings of the values in the columns

corrections = {
    'Karapakam': 'Karapakkam',
    'Ana Nagar': 'Anna Nagar',
    'Ann Nagar': 'Anna Nagar',
    'Adyr': 'Adyar',
    'Velchery': 'Velachery',
    'Chrompt': 'Chrompet',
    'Chrmpet': 'Chrompet',
    'Chormpet': 'Chrompet',
    'KKNagar': 'KK Nagar',
    'TNagar': 'T Nagar',
}

corrections_SALE_COND = {
    'PartiaLl': 'Partial',
    'Partiall': 'Partial',
    'Ab Normal': 'AbNormal',
    'Adj Land': 'AdjLand'
}

corrections_PARK_FACIL = {
    'Noo': 'No'
}

corrections_BUILDTYPE = {
    'Comercial': 'Commercial',
    'Other' : 'Others'
}

corrections_UTILITY_AVAIL = {
    'AllPub': 'All Pub',
    'NoSewr ' : 'NoSeWa'
}

corrections_STREET = {
    'Pavd': 'Paved',
    'NoAccess' : 'No Access'
}

df['AREA'] = df['AREA'].replace(corrections)
df['SALE_COND'] = df['SALE_COND'].replace(corrections_SALE_COND)
df['PARK_FACIL'] = df['PARK_FACIL'].replace(corrections_PARK_FACIL)
df['BUILDTYPE'] = df['BUILDTYPE'].replace(corrections_BUILDTYPE)
df['UTILITY_AVAIL'] = df['UTILITY_AVAIL'].replace(corrections_UTILITY_AVAIL)
df['STREET'] = df['STREET'].replace(corrections_STREET)
```

```python
#To check for null values in the columns

df.isnull().sum() #--> null values are there
```

```python
for idx,row in df[df['N_BEDROOM'].isnull()].iterrows():
  match = df[(df['AREA']== "Anna Nagar") & (df['N_ROOM']==4) & (df['INT_SQFT']>=1500) & (df['DIST_MAINROAD']==181)]
  if not match.empty:
    df.at[idx,"N_BEDROOM"] = match.iloc[0]["N_BEDROOM"]
```

```python
df['N_BATHROOM'] = df.apply(lambda row: row["N_BEDROOM"] if pd.isnull(row["N_BATHROOM"]) else row["N_BATHROOM"], axis=1)
```

```python
df['QS_OVERALL'] = df.apply(lambda row: (row["QS_ROOMS"] + row["QS_BEDROOM"] + row["QS_BATHROOM"])/3 if pd.isnull(row["QS_OVERALL"]) else row["QS_OVERALL"], axis=1)
```

```python
#For time related column we are using to_datetime() to change the datatype to datetime64[ns]

df["DATE_SALE"]=pd.to_datetime(df["DATE_SALE"])
df["DATE_BUILD"]=pd.to_datetime(df["DATE_BUILD"])

#-------------------------------------------------------------

df["DB_Year"]=pd.to_datetime(df["DATE_BUILD"]).dt.year
df["DB_Month"]=pd.to_datetime(df["DATE_BUILD"]).dt.month
df["DB_Day"]=pd.to_datetime(df["DATE_BUILD"]).dt.day
df["DS_Year"]=pd.to_datetime(df["DATE_SALE"]).dt.year
df["DS_Month"]=pd.to_datetime(df["DATE_SALE"]).dt.month
df["DS_Day"]=pd.to_datetime(df["DATE_SALE"]).dt.day
```

```python
#From the correlation matrix we can observe that "PRT_ID","DATE_BUILD","DATE_SALE" columns have no great relation to the data
 
df = df.drop(["PRT_ID","DATE_BUILD","DATE_SALE"],axis="columns")
```

```python
#Separating  the dependent(Target) and independant columns

x = df.drop("SALES_PRICE",axis=1) #independent
y = df["SALES_PRICE"] #dependent
```

```python
#Spliting the data into 90% for training and 10% for testing

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.1)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(x_train,y_train)
```

```python
y_predict = model.predict(x_test)
print("MAE : ",mean_absolute_error(y_test,y_predict))
print("MSE : ",mean_squared_error(y_test,y_predict))
print("r2 : ",r2_score(y_test,y_predict))
```

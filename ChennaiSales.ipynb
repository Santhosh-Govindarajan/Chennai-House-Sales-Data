```python
from scipy import stats
```

```python
#Continues VS Continues
```

```python
def two_sample(d1,d2):
    t = 0
    f = 0
    for i in range(31):
        sample1 = d1.sample(frac=0.05)
        sample2 = d1.sample(frac=0.05)
        t_test,p_value = stats.ttest_ind(sample1,sample2)
        if p_value < 0.02: #no connection
            f = f+1   
        else:              #if p_value > 0.02 connection is there
            t = t+1
        if f>t:
            return False
        else:
            return True
```

```python
two_sample(df["COMMIS"],df["SALES_PRICE"])
```

```python
#Category Vs Category
#[0] = stats value(f_value)      [1] = probability value(p_value)  [2] = degree of freedom  [3] = Expected
```

```python
def chi(d1,d2):
    return True if stats.chi2_contingency(pd.crosstab(d1,d2))[1] > 0.02 else False
```

```python
chi(df["AREA"],df["BUILDTYPE"])
```

```python
#Continues Vs Category
#d1 - Continues
#d2 - Category
```

```python
def f_test(d1,d2):
    group = df[d2].unique()
    data = {}
    for i in group:
        data[i] = df[d1][df[d2]==i]
    f_value,p_value = stats.f_oneway(*[i for i in data.values()])    #Anava test
    if p_value < 0.02:
        return False
    else:
        return True
```

```python
f_test("AREA","BUILDTYPE")
```

```python
final={}
for i in df.columns:
    final[i]={}
    for j in df.columns:
        if (i in continues) & (j in continues):
            result = two_sample(df[i],df[j])
        elif (i in continues) & (j in category):
            result = f_test(i,j)
        elif (i in category) & (j in continues):
            result = f_test(j,i)
        elif (i in category) & (j in category):
            result = chi(df[i],df[j])
        if result:
            final[i][j] = 1
        else:
            final[i][j] = 0
```
